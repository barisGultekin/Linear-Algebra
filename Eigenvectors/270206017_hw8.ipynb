```python
from fractions import Fraction

def fractionalMatrix(M):
    res = []
    for r in range(len(M)):
        resRow = []
        for c in range(len(M[r])):
            resRow.append(Fraction(M[r][c]))
        res.append(resRow)
    return res

def nPrint(M):
    s = "["
    for r in range(len(M)):
        s+="["
        for c in range(len(M[r])):
            t = str(M[r][c])
            s+=t
            if (c!=len(M[r])-1):
                s+=", "

        if (r!=len(M)-1):
            s+="], \n"
        else:
            s+="]"
    s +="]"
    print(s)
    return True
```

```python
def MatrixDim(A):
    #
    # This function is supposed to return the dimensions of matrix A
    #
    M265rows = len(A)
    if (M265rows <= 0):
        return[-1,-1]
    M265cols = len(A[0])
    # we will check that this is indeed matrix i.e. number of cols in every row is the same
    for i in range(M265rows):
        if (len(A[i]) != M265cols):
            return [-1,-1]
    return [M265rows, M265cols]  



def MatrixMult(A,B):
    #
    # This function should multiply the matrices A and B
    # start by creating an empty array
    [rowsA,colsA] = MatrixDim(A)
    [rowsB,colsB] = MatrixDim(B)
    if ((colsA != rowsB) and (colsA > 0)):
        # if matrix multiplication cannot be done return an empty array
        return []
        
    M265Result = [[0 for k in range(colsB)] for r in range(rowsA)]
    for rw in range(rowsA):
        for cl in range(colsB):
            tmp =0
            for k in range(colsA):
                tmp = tmp + A[rw][k]*B[k][cl]
            M265Result[rw][cl]=tmp
    return M265Result
```

```python
def zerosMatrix(rows, cols):
    M = []
    while len(M) < rows:
        M.append([])
        while len(M[-1]) < cols:
            M[-1].append(0.0)
    return fractionalMatrix(M)

def copyMatrix(M):
    rows = len(M)
    cols = len(M[0])
    MC = zerosMatrix(rows, cols)

    for i in range(rows):
        for j in range(cols):
            MC[i][j] = M[i][j]

    return fractionalMatrix(MC)

def identityMatrix(n):
    IdM = zerosMatrix(n, n)
    for i in range(n):
        IdM[i][i] = 1.0
    return fractionalMatrix(IdM)

def matrixSubtraction(A, B):
    rowsA = len(A)
    colsA = len(A[0])
    rowsB = len(B)
    colsB = len(B[0])
    if rowsA != rowsB or colsA != colsB:
        print("Matrix is not symetric.")
        return False
    C = zerosMatrix(rowsA, colsB)
    for i in range(rowsA):
        for j in range(colsB):
            C[i][j] = A[i][j] - B[i][j]
    return C

def matrixMultConstant(A:list,c):
    res = [[0 for _ in range(len(A))] for _ in range(len(A[0]))]
    for i in range(len(A)):
        for j in range(len(A[0])):
            res[i][j] = A[i][j]*c
    return res
```

<h2>Linear dependence</h2>

Consider the sequence $\vec{v}_0 = \vec{v}$, $\vec{v}_1 = M\vec{v}_0\dots$ defined recursively $$\vec{v}_{i+1} = M\vec{v}_i$$. Since all vectors lie in the same $d$-dimensional vector space then there is a smallest integer $k$ such that the set $\vec{v}_0,\dots\vec{v}_{k}$ is linearly dependent but any set  $\vec{v}_0,\dots\vec{v}_{j}$ with $j<k$ is linearly independent. With that in mind it is necessarily the case that in the equation $$\alpha_0\vec{v}_0 + \alpha_1\vec{v}_1+\cdots+\alpha_{k-1}\vec{v}_{k-1}+\alpha_{k}\vec{v}_k =\vec{0}$$ the coefficient $\alpha_k\ne 0$.
